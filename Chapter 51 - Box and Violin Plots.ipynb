```python
data1= randn(100)
data2 =randn(100)
```

```python
sns.boxplot(data=[data1,data2])
 # two data  in one chart
```

```python
# black dot on top of blue bar and  same at the bottom of orange bar is called "outlier point"
# including thoes into range of whisker
sns.boxplot(data=[data1,data2],whis=np.inf)
```

```python
sns.boxplot(x=data2,whis=np.inf)
```
