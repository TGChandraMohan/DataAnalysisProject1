```python
# import  plotting
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#this alow us to see our plots in python notebook
%matplotlib inline
```

```python
# making a data set
```

```python
dataset  = randn(100)
```

```python
sns.distplot(dataset,bins=25,hist_kws=dict(edgecolor="black", linewidth=2))
#   'distplot'allows us to combine plots
# 'hist_kws=dict(edgecolor="black", linewidth=2)' this function allow us edger color in seaborn
```

```python
#  by default 'distplot' combines KDE plot with a histogram but we can also choose whetehr its a rug plot or KDE or histogram
sns.distplot(dataset,bins=25,rug=True,hist=False)
```

```python
# controlling specific plot within the combined plot, we can do that by passsing dictionaries
```

```python
sns.distplot(dataset,bins=25,
            kde_kws={'color':'indianred','label':'KDE Plot'},
            hist_kws={'color':'blue','label':'HIST','edgecolor':'black', 'linewidth':2})
plt.legend()
```
