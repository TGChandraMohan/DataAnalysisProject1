```python
points =np.arange(-5,5,0.05)
```

```python
dx,dy = np.meshgrid(points,points) # CReating a Grid, with points as Vector
```

```python
dx  # array from -5 to  + 5 row vise (  horizntal)
```

```python
dy # array from -5 to +5  column vise ( vertrically)
```

```python
# Example 2 
Evaluvating function
```

```python
# calculating  Trigonometry Sign
z=(np.sin(dx)+ np.sin(dy))
z
```

```python
# Example 3 - Ploting  graph
```

```python
plt.imshow(z)
```

```python
plt.imshow(z)
plt.colorbar()

plt.title ('plot of sin (x)+ sin (y)')
```
