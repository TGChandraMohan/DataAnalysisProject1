```python
# how to see both  histogram on the same plot
```

```python
plt.hist(dataset1,density=True,color='yellow',alpha=0.5,bins=20, edgecolor='k')
plt.hist(dataset2,density=True,alpha=0.5,bins=20,edgecolor='k')
# since both data set have  differnt data point 80 and 100 use  "normad= True" function to  normalise
# "alpha" function is for transperancy and  level of transperancy is 0.5
# by default no of bins is 10 but we can alter usnig "bins"function
```
