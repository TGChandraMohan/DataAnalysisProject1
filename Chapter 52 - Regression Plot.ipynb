```python
import numpy as np
import pandas as pd
from numpy.random import randn


from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns


%matplotlib inline
```

```python
# seborn as some sample data set, we using one among thoes
tips =sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.lmplot('total_bill','tip',tips)
```

```python
# altering  feater of  scater plot , like color and width size and  level of order
sns.lmplot('total_bill','tip',tips,order=4,
          scatter_kws={'marker':'o','color':'indianred'},
          line_kws={'linewidth':0.5,'color':'blue'})
```

```python
# sample 2 to check level of order
sns.lmplot('total_bill','tip',tips,order=10,
          scatter_kws={'marker':'o','color':'indianred'},
          line_kws={'linewidth':0.5,'color':'blue'})
```

```python
# not fitting a regression at all
```

```python
sns.lmplot('total_bill','tip', tips, fit_reg=False)
# fit_ reg= False  will remove regresssion line
```

```python
# lmplot works on discrete valiable such as tips %
```

```python
tips.head()
```

```python
# adding a column tips % insde tips data set
tips['tip_pect']=100*(tips['tip']/tips['total_bill'])
```

```python
tips.head()
```

```python
# Ploting  tips %
sns.lmplot(x='size',y='tip_pect',data= tips)
```

```python
sns.lmplot('size','tip_pect',tips)
```

```python
# adding a jitter, check wiki to know what is jitter
```

```python
sns.lmplot('size','tip_pect',tips,x_jitter=0.1)
```

```python
# estimating tendency of each bin
```

```python
sns.lmplot('size','tip_pect',tips,x_estimator=np.mean)
```

```python
# hue facet
```

```python
sns.lmplot('total_bill','tip_pect',tips,hue='sex',markers=['X','o'])
```

```python
# data between day and amount of tip
```

```python
sns.lmplot('total_bill','tip_pect',tips,hue='day')
```

```python
# local regression model fitting
```

```python
sns.lmplot('total_bill','tip_pect',tips,lowess=True, line_kws={'color':'black'})

# Lowess = local regression
```

```python
# Lower level fiunction
sns.regplot('total_bill','tip_pect',tips, line_kws={'color':'red',},marker='x')
```

```python
# subplots
```

```python
fig,(axis1,axis2)=plt.subplots(1,2,sharey=True)

sns.regplot('total_bill','tip_pect',tips,ax=axis1)
sns.violinplot(tips['tip_pect'],tips['size'],color='red',ax=axis2)
```
